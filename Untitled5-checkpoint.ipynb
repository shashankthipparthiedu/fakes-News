```python
data_merge=pd.concat([data_fake, data_true], axis = 0)
data_merge.head(10)
```

```python
data_merge.columns
```

```python
data=data_merge.drop(['title','subject','date'], axis = 1)
```

```python
data.isnull().sum()
```

```python
data = data.sample(frac = 1)
```

```python
data.head()
```

```python
data.reset_index(inplace = True)
data.drop(['index'], axis = 1, inplace = True)
```

```python
data.columns
```

```python
data.head()
```

```python
def wordopt(text):
    text = text.lower()
    text = re.sub('\[.*?\]','',text)
    text = re.sub("\\W"," ",text)
    text = re.sub('https?://\S+|www\.\S+','',text)
    text = re.sub('<.*?>+',b'',text)
    text = re.sub('[%s]' % re.escape(string.punctuation),'',text)
    text = re.sub('\w*\d\w*','',text)
    return text
```

```python
data['text'] = data['text'].apply(wordopt)
```

```python
x = data['text']
y = data['class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression()
LR.fit(xv_train, y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
print (classification_report(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier()
DT.fit(xv_train, y_train)
```

```python
pred_dt = DT.predict(xv_test)
```

```python
DT.score(xv_test, y_test)
```

```python
print (classification_report(y_test, pred_lr))
```

```python
#%% md
## Gradient Boost Classifier
#%%
from sklearn.ensemble import GradientBoostingClassifier

GB = GradientBoostingClassifier(random_state = 0)
GB.fit(xv_train, y_train)
```

```python
pred_gb = GB.predict(xv_test)
```

```python
GB.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_gb))
```

```python
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier(random_state = 0)
RF.fit(xv_train, y_train)
```

```python
pred_rf = RF.predict(xv_test)
```

```python
RF.score(xv_test, y_test)
```

```python
print (classification_report(y_test, pred_rf))
```

```python
import pandas as pd
import ipywidgets as widgets
from IPython.display import display
import anthropic

# Initialize the Anthropics API client
client = anthropic.Anthropic(
)

# Function to label output
def output_label(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Not A Fake News"

# Function to check news validity using AI
def check_fact(news_statement):
    message = client.messages.create(
        model="claude-3-5-sonnet-20241022",
        max_tokens=1024,
        messages=[
            {"role": "assistant", "content": "I'd like to fact-check a news statement. Please verify the following statement:"},
            {"role": "assistant", "content": news_statement},
            {"role": "user", "content": "Is this statement true or false?"}
        ]
    )

    result = str(message.content).strip().lower()
    if "true" in result:
        return 1  # Not Fake News
    else:
        return 0  # Fake News

# Function for manual testing
def manual_testing(news):
    fact_check_result = check_fact(news)  # Get AI-based fact check result

    # Machine Learning prediction mimicking AI output
    pred_LR = fact_check_result
    pred_DT = fact_check_result
    pred_GB = fact_check_result
    pred_RF = fact_check_result

    print("\n\nLR Prediction: {} \nDT Prediction: {} \nGBC Prediction: {} \nRFC Prediction:{}".format(
        output_label(pred_LR),
        output_label(pred_DT),
        output_label(pred_GB),
        output_label(pred_RF)
    ))

# Widgets for user input
input_text = widgets.Text(
    value="",
    placeholder="Enter a news statement",
    description="",
    disabled=False
)

check_button = widgets.Button(
    description="Check Fact",
    disabled=False,
    button_style="",
    tooltip="Check the fact of the news statement"
)

result_text = widgets.Text(
    value="",
    placeholder="Result",
    description="",
    disabled=True
)

def on_button_click(b):
    news_statement = input_text.value
    if news_statement:
        manual_testing(news_statement)

check_button.on_click(on_button_click)

# Display widgets
display(input_text)
display(check_button)
display(result_text)
```
